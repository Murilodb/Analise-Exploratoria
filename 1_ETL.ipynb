```python
#fazendo correção da coluna "tipo_de_cliene"
df.rename(columns={'Tipo de Cliene': 'Tipo de Cliente'}, inplace=True)
df.head()
```

```python
df2= df.copy()
```

```python
#alterando nome das colunas padronizando como minusculas
df2.columns = df2.columns.str.lower()
df2.head(1)
```

```python
#tirando caracteres invalodos das colunas
df2.columns = df2.columns.str.replace('($)','')


df2.head(1)
```

```python
#removendo espaços no inicio e fim do nome da coluna
df2.columns = df2.columns.str.strip()
```

```python
#substituido espaços por underline
df2.columns = df2.columns.str.replace(' ','_')
df2
```

```python
#removendo acentuação nos nomes da colunas
lenghtColumns = len(df2.columns)

colunas_formatadas = []

for col in range(lenghtColumns):
    colunas_formatadas.append(unidecode(df2.columns[col]))


df2.columns = colunas_formatadas

df2.head()
```

```python
agg2 = df2.groupby('tipo_de_cliente').size().reset_index(name='quantidade')
agg1 = df2.groupby('metodo_de_pagamento').size().reset_index(name='quantidade')
agg3 = df2.groupby('genero').size().reset_index(name='quantidade')
agg4 = df2.groupby('estado_civil').size().reset_index(name='quantidade')

agg1
```
